Missing Values are imputed by ForwardFill and Backward Fill

```python
import pandas as pd

# Assuming df is your DataFrame with NaN values in the ID column
# df = ...

# Identify numeric ID columns
numeric_id_cols = [col for col in df.columns if df[col].dtype == 'int' or df[col].dtype == 'float']

# Impute missing values for numeric ID columns
for col in numeric_id_cols:
    df[col] = df[col].fillna(method='ffill')  # Forward fill missing values with the previous value

# If there are still missing values after forward fill, you can also use backward fill
for col in numeric_id_cols:
    df[col] = df[col].fillna(method='bfill')  # Backward fill missing values with the next value
```

```python
def encoding(dataset, categorical):
    """
    Function to automate the process of encoding the categorical data
    :param dataset: Dataframe
    :param categorical: List of categorical columns
    :return: Dataframe
    """
    for i in categorical:
        dataset[i] = dataset[i].astype('category')
        dataset[i] = dataset[i].cat.codes
    return dataset
encoding(df, categorical)
```

```python
from sklearn.preprocessing import PowerTransformer
```

Since Box Cox Transformation can be applied only to positive dataset we can use the Yeo-Johnson transformation, which is an extension of the Box-Cox transformation and supports both positive and negative values.

```python
# Initialize PowerTransformer object
power_transformer = PowerTransformer(method='yeo-johnson', standardize=False)

# Fit and transform the data
transformed_data = power_transformer.fit_transform(df)

# Print the transformed data
print(transformed_data)
```
